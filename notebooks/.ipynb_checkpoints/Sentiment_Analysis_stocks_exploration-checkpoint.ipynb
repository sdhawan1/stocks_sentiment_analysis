```python
'''
    how much stock data do I have on my hands here....?
    search for the common stocks and figure out how many articles per date.
    I can then cross-reference with stock price and get a decent dataset going (maybe).
'''

import datetime as dt
from bs4 import BeautifulSoup
import requests

## first step: collect the dates from the html

def collect_total_items():
    #1. collect all the rss feeds.
    resp = requests.get("https://www.cnbc.com/rss-feeds/")
    soup = BeautifulSoup(resp.content, "lxml")
    div = soup.find("div", attrs={"class" :  "wildcard"})
    links = []
    total_items = []
    i = div.find_all("a")
    #print (i[0].attrs["href"])
    for k in i:
        links.append(k.attrs["href"])

    #2. Crawl each rss feed and collect all articles.
    for link in links:
        print ("https:"+link) #prints the rss feed link.

        test = requests.get("https:"+link)
        test_soup = BeautifulSoup(test.content,"lxml")
        #item is for each RSS link.
        items = test_soup.findAll("item")
        total_items.append(items)
    return total_items

#get all the metadata (don't just filter out the title & description yet...)
all_md = collect_total_items()
print(all_md[0][:4])
```

```python
## TODO: REPLACE THE ORIGINAL CRAWLER CODE WITH THIS.
### (gathers the title and the date for every article, not just the title.)

#next step: gather the date + title for every article.
dictionaries = []
d1 = {}
for i in range(len(all_md)):
    for j in range(len(all_md[i])):
        try:
            dictionary = d1.copy()
            dictionary["title"] = all_md[i][j].title.text.lower()
            dictionary["description"] = all_md[i][j].description.text.lower()
            dictionary["date"] = all_md[i][j].pubdate.text
            #print (dictionary)
            dictionaries.append(dictionary)
        except Exception as e:
            print(type(e))
            print(e)

print(dictionaries[:10])
```

```python
#analyze the result: count the spread of dates...
import matplotlib.pyplot as plt
print(len(dictionaries))

#count spread of dates!
art_dates = []
for article in dictionaries:
    #metadata_date
    md_date = article['date']
    #interpretation...
    d = dt.strptime(md_date, "%a, %d %b %Y %X %Z")
    month_day = str(d.month)+'_'+str(d.day)
    art_dates.append(month_day)

print(art_dates[:20])
days = [int(d.split('_')[1]) for d in art_dates]
days = [d for d in days if d <= 12]


n, bins, patches = plt.hist(days, 12)
f = plt.plot(bins)
plt.show()

## results: there are around 3 days of solid data.
## if you had 10-15 stock tickers, that's around 30-45 data points, which isn't the best...

#try to get the archived rss feed?
```
